###刪除UnitPrice<0 或是UnitPrice > 3000的銷售資料

```python
data0 = data.iloc[1284117:1344566] #設定1101-1130的資料為變數data0
data0
```

```python
data0[(data0.UnitPrice<0)|(data0.UnitPrice>3000)] 
#測試網路查詢資料，找出1101-1130中"UnitPrice為<0"『或』">3000的資料"
#data0[(data0.UnitPrice<0)&(data0.UnitPrice>3000)] 找出..".."『和』".."
```

```python
#data0[~(data0.UnitPrice<0)&~(data0.UnitPrice>3000)] 
#詢問同學及網路爬文得知,「 ~ 」 為反義詞 上述語法得出：『data0中「不是」<0『也』「>3000」的資料』
data2 = data0[~(data0.UnitPrice<0)&~(data0.UnitPrice>3000)] 
data2
```

###抓出CustomerID,StockCode,Description並重新存成一個新的dataframe

```python
#抓出題目需要之資料
data3 = data2[['CustomerID','StockCode','Description']]
data3
```

```python
#儲存成一個新的.csv
data3.to_csv('New_Data.csv',encoding='utf8')
```
